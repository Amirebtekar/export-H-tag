```python
import requests
from bs4 import BeautifulSoup
import re

def striphtml(data):
    p = re.compile(r'<.*?>')
    return p.sub('',data)

urls = ['https://besazobechin.com/Blog/267/50-%D9%85%D8%AF%D9%84-%DA%A9%D8%AA%D8%A7%D8%A8%D8%AE%D8%A7%D9%86%D9%87-%D8%AF%DB%8C%D9%88%D8%A7%D8%B1%DB%8C-%D8%B3%D8%A7%D8%AF%D9%87-%D9%84%D8%A7%DA%A9%DA%86%D8%B1%DB%8C-%D8%AC%D8%AF%DB%8C%D8%AF-2022',
        'https://arga-mag.com/104120/%D9%85%D8%AF%D9%84-%DA%A9%D8%AA%D8%A7%D8%A8%D8%AE%D8%A7%D9%86%D9%87',
        'https://www.beautyhome.ir/bed-dresser/51069/%D9%85%D8%AF%D9%84-%DA%A9%D8%AA%D8%A7%D8%A8%D8%AE%D8%A7%D9%86%D9%87-2',
        'https://www.digikala.com/search/category-home-library/',
        'https://moblomiz.com/Blog/newLibrary',
        'https://etude.design/%DA%A9%D8%AA%D8%A7%D8%A8%D8%AE%D8%A7%D9%86%D9%87-%D8%AF%DB%8C%D9%88%D8%A7%D8%B1%DB%8C/',
        'https://asaschin.com/%D9%85%D8%AF%D9%84-%DA%A9%D8%AA%D8%A7%D8%A8%D8%AE%D8%A7%D9%86%D9%87-%D8%AF%DB%8C%D9%88%D8%A7%D8%B1%DB%8C-%D9%85%D8%AF%D8%B1%D9%86/',
        'https://drchoob.ir/product/bookshelf-f30/',
        'https://mag.sazokar.com/newest-bookshelf-models/',
       ]
for x1 in urls:
    
    url = x1
    req = requests.get(url)
    soup = BeautifulSoup(req.content, 'html.parser')

    a = soup.find_all(['h1','h2','h3','h4'])

    for i in a:
        x = striphtml(str(i))
        print(x)
```
