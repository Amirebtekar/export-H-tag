```python
import requests
from bs4 import BeautifulSoup
import re

def striphtml(data):
    p = re.compile(r'<.*?>')
    return p.sub('',data)

urls = []
f = open('url.txt','r').read().split('\n')

for i in f:
    urls.append(i)

for x1 in urls:
    
    url = x1
    req = requests.get(url)
    soup = BeautifulSoup(req.content, 'html.parser')

    a = soup.find_all(['h1','h2','h3','h4'])

    for i in a:
        x = striphtml(str(i))
        print(x)
```
